## Importing the required Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.layers import SimpleRNN, LSTM, Dense, Input
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
```

### Generating a difficult sequence

```python
data = np.sin((0.2 * np.arange(300)) ** 2)
plt.figure(figsize = (12, 7))
plt.plot(data)
plt.show()
```

```python
# Generating the TimeSeries forecasting Data
Sequence_Length = 15
Dimension = 1
TimeSeries = []
Forecast = []

for sequence in range(len(data) - Sequence_Length):
    TimeSeries.append( data[ sequence : sequence + Sequence_Length ] )
    Forecast.append( data[ sequence + Sequence_Length ] )

TimeSeries = np.array(TimeSeries)
Forecast = np.array(Forecast)
```

```python
## Building AR(Auto Regressive) Model
inputShape = Input(shape = (Sequence_Length, ))
FinalLayer = Dense(1)(inputShape)

ARModel = Model(inputShape, FinalLayer)
ARModel.compile(loss = "mse", optimizer = Adam(lr = 0.01))
```

```python
ARModel.summary()
```

```python
limit = TimeSeries.shape[0] // 2
trainedModel = ARModel.fit(TimeSeries[: limit],
            Forecast[: limit],
            epochs = 300,
            validation_data = (TimeSeries[limit : ], Forecast[limit : ])
           )
```

```python
plt.figure(figsize = (12, 7))
plt.plot(trainedModel.history["loss"], label = "loss" )
plt.plot(trainedModel.history["val_loss"], label = "Val_loss" )
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
## Making predictions from the Model on the subset of the dataset
predictions = ARModel.predict(TimeSeries[limit: ])
```

```python
TimeSeries[limit:].shape
```

```python
plt.figure(figsize = (12, 7))
plt.plot(Forecast[limit : ], label = "Actual Values" )
plt.plot(predictions, label = "Predicted Values" )
# plt.xlabel("Epochs")
# plt.ylabel("Loss")
plt.legend()
plt.show()
```

***The above figure shows that if we have a Difficult Sequence, then we have horrible result using AR Linear Model***

```python
## Using Right Forecast approach which may perform more worse
actualLabels = Forecast[limit : ]
predictions = []
lastValue = TimeSeries[limit]

while len(predictions) < len(actualLabels):
    temp = ARModel.predict(lastValue.reshape(1, -1))[0, 0]
    predictions.append(temp)
    
    lastValue = np.roll(lastValue, -1)
    lastValue[-1] = temp
```

```python
plt.figure(figsize = (12, 7))
plt.plot(actualLabels, label = "Actual Values" )
plt.plot(predictions, label = "Predicted Values" )
# plt.xlabel("Epochs")
# plt.ylabel("Loss")
plt.legend()
plt.show()
```

## Building RNN Model

```python
## Building Simple RNN Model
inputShape = Input(shape = (Sequence_Length, ))
LayerStack = SimpleRNN(5)(inputShape)
LayerStack = Dense(1)(LayerStack)

ARModel = Model(inputShape, LayerStack)
ARModel.compile(loss = "mse", optimizer = Adam(lr = 0.01))

limit = TimeSeries.shape[0] // 2
trainedModel = ARModel.fit(TimeSeries[: limit],
            Forecast[: limit],
            epochs = 300,
            validation_data = (TimeSeries[limit : ], Forecast[limit : ])
           )
```
