```python
# Importing the required Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_s_curve
from sklearn.manifold import Isomap
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Generating the Non-Linear dataset
X, color = make_s_curve(500)
```

```python
X.shape
```

```python
X[:, :0].shape
```

```python
# Plotting the data
fig = plt.figure(figsize = (15, 15))
ax = fig.add_subplot(projection = "3d")
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c = color, s = 50, cmap = plt.cm.Accent_r)
plt.show()
```

```python
# Generating the IsoMapping
mapper = Isomap(n_neighbors = 7, n_components = 2)
```

```python
new_X = mapper.fit_transform(X)
```

```python
new_X.shape
```

```python
# Plotting the reduced dimensions data in 2-D
fig = plt.figure(figsize = (15, 9))
plt.scatter(X[:, 0], X[:, 1], c = color, s = 50, cmap = plt.cm.Accent_r)
plt.show()
```
