```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
```

## Data Processing

```python
data = pd.read_csv("https://drive.google.com/uc?id=1bW1J33xCyAf0glj72_OLiEMAm2Z9ddzx")
```

```python
data
```

# Plotting Dendrograms

```python
columns = data.iloc[:, [3, 4]].values

plt.figure(figsize = (15, 35))
dend = dendrogram(linkage(columns, method = 'complete'), orientation = "right")
```

## Creating Model of Agglomerative Clustering

```python
Hierarchical_Clustering_Model = AgglomerativeClustering(n_clusters = 5, affinity='euclidean', linkage='complete')
Hierarchical_Clustering_Model.fit(columns)
```

```python
labels = Hierarchical_Clustering_Model.labels_
print(labels)
```

```python
plt.figure(figsize = (15, 7))
plt.scatter(columns[labels == 0, 0], columns[labels == 0, 1], s = 100, marker = 'D', color = 'brown')
plt.scatter(columns[labels == 1, 0], columns[labels == 1, 1], s = 100, marker = '1', color = 'red')
plt.scatter(columns[labels == 2, 0], columns[labels == 2, 1], s = 100, marker = 'H', color = 'grey')
plt.scatter(columns[labels == 3, 0], columns[labels == 3, 1], s = 100, marker = '^', color = 'orange')
plt.scatter(columns[labels == 4, 0], columns[labels == 4, 1], s = 100, marker = '*', color = 'green')
plt.show()
```
