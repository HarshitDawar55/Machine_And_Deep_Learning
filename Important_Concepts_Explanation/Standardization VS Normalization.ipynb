```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
data = np.random.randint(1, 100, size = (100, 2))
```

```python
plt.figure(figsize = (15, 7))
plt.scatter(data[:, 0], data[:, 1], c = "r", marker = "*", s = 100)
plt.plot(data[:, 0], data[:, 1])
```

```python
scaledData = StandardScaler().fit_transform(data)
```

```python
plt.figure(figsize = (15, 7))
plt.scatter(scaledData[:, 0], scaledData[:, 1],= c = "r", marker = "*", s = 100)
# plt.xlim(0, 1)
# plt.ylim(0, 1)
plt.plot(scaledData[:, 0], scaledData[:, 1])
```

```python
scaledData = MinMaxScaler().fit_transform(data)
```

```python
plt.figure(figsize = (15, 7))
plt.scatter(scaledData[:, 0], scaledData[:, 1], c = "r", marker = "*", s = 100)
# plt.xlim(0, 1)
# plt.ylim(0, 1)
plt.plot(scaledData[:, 0], scaledData[:, 1])
```
