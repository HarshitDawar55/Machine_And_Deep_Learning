```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
data = np.random.randint(1, 100, size = (10, 10))
```

```python
data
```

## Standardized Data

```python
scaledData = StandardScaler().fit_transform(data)
```

```python
scaledData
```

## Normalized Data

```python
scaledData = MinMaxScaler().fit_transform(data)
```

```python
scaledData
```
