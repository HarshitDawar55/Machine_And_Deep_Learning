```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use("seaborn-poster")
plt.figure(figsize = (15, 7))
plt.plot([0, 1, 2, 3, 4, 5, 6], [18, 19, 20, 21, 22, 23, 24],  c = "orange", linestyle = "solid", linewidth = 5)
plt.scatter([2, 3, 5, 7, 9, 15, 23], [20, 21, 15, 17, 23, 5, 3], c = "red", marker = "*")
plt.xlim(0, 25)
plt.ylim(0, 25)
plt.title("High Bias Graph")
plt.show()
```

```python
plt.style.use("seaborn-poster")
plt.figure(figsize = (15, 7))
plt.plot([2, 3, 5, 7, 9, 15, 23], [20, 21, 15, 17, 23, 5, 3], c = "lightgreen", linestyle = "solid", linewidth = 5)
plt.scatter([2, 3, 5, 7, 9, 15, 23], [20, 21, 15, 17, 23, 5, 3], c = "red", marker = "*", s = 700)
plt.xlim(0, 25)
plt.ylim(0, 25)
plt.title("High Variance Graph")
plt.show()
```
