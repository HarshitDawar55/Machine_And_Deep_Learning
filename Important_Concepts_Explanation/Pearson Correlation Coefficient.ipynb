```python
# Importing the required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Generating synthetic Data for different values of the Pearson Correlation coefficiant!

### Part 1: Negatively Correlated

```python
x = np.arange(1, 15, 1)
y = np.arange(15, 1, -1)
```

```python
plt.figure(figsize = (12, 7))
plt.scatter(x, y, marker = "*", s = 500, c = "r")
plt.plot(x, y, linewidth = 3, linestyle = "solid", c = "lightgreen")
plt.title("Highly Negatively Correlated Features Example")
plt.show()
```

```python
# Value of the Pearson Correlation between x & y
np.corrcoef(x, y)[0][1]
```

### Part 2: Correlation between -1 & 1

```python
x = np.random.randint(1, 15, size = (1, 15))
y = np.random.randint(1, 15, size = (1, 15))
```

```python
x.shape, y.shape
```

```python
plt.figure(figsize = (12, 7))
plt.scatter(x, y, marker = "*", s = 500, c = "r")
plt.title("Randomly Correlated Features Example")
plt.show()
```

```python
# Value of the Pearson Correlation between x & y
np.corrcoef(x, y)[0, 1]
```

### Part 3: Positively Correlated!

```python
x = np.arange(1, 15, 1)
y = np.arange(1, 15, 1)
```

```python
plt.figure(figsize = (12, 7))
plt.scatter(x, y, marker = "*", s = 500, c = "r")
plt.plot(x, y, linewidth = 3, linestyle = "solid", c = "lightgreen")
plt.title("Highly Positively Correlated Features Example")
plt.show()
```

```python
np.corrcoef(x, y)[0, 1]
```
