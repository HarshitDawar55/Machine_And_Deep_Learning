```python
# Importing the required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Generating synthetic Data for different values of the Pearson Correlation coefficiant!

### Part 1 Negatively Correlated

```python
x = np.arange(1, 15, 1)
y = np.arange(15, 1, -1)
```

```python
plt.figure(figsize = (12, 7))
plt.scatter(x, y, marker = "*", s = 500, c = "r")
plt.plot(x, y, linewidth = 3, linestyle = "solid", c = 'sa')
plt.show()
```

```python
np.corrcoef(x, y)
```
